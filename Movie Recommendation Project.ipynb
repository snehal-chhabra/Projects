### Recommendaion Engine based on Content-Based Filtering 

```python
dataset = pd.read_csv('movies_dataset.csv')
dataset
```

```python
movies_df = dataset.copy()
```

```python
movies_df.drop('Unnamed: 0', axis = 1, inplace = True)
movies_df
```

```python
movies_df.info()
```

```python
movies_df['tags'][3]
```

```python
#How to recommend movie on the basis of description/tags/comments/genres/remarks
#Bag of words creation is required --> 
#WORD COUNT VECTORIZER --> It will take all the words
#TFIDF VECTORIZER --> It only takes important words, ignores verbs, adverbs, etc
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
overview_matrix = tfidf.fit_transform(movies_df['tags'])
```

```python
overview_matrix.shape
```

```python
count_array = overview_matrix.toarray()
count_array
```

```python
#if you want to see the bag of words..
overview_matrix2 = pd.DataFrame(data =count_array, columns  = tfidf.get_feature_names_out())
overview_matrix2
```

```python
# we will apply the cosine similarity on the count_array or on overview_matrix2
#-- which is the array format of Bag of words
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = pd.DataFrame(cosine_similarity(overview_matrix2))
similarity_matrix
```

```python
similarity_score = list(enumerate(similarity_matrix[3]))
similarity_score
```

```python
similarity_score = sorted(enumerate(similarity_matrix[3]), key = lambda x: x[1], reverse = True)
similarity_score
```

```python
#Basic mapping index of movie
mapping = pd.Series(movies_df.index, index = movies_df['title'])
mapping
```

```python
movie_index = mapping['Spectre']
movie_index
```

```python
#write a function --> pass a movie name --> get recommendation
def recommend_movie(movie_input):
    movie_index = mapping[movie_input]
    #it takes the similarity value of other movies
    similarity_score = list(enumerate(similarity_matrix[movie_index]))
    similarity_score = sorted(similarity_score, key = lambda x: x[1], reverse = True)
    #we want to recommend 10 movies
    similarity_score = similarity_score[1:11]
    movie_ind = [i[0] for i in similarity_score]
    list_recommend = movies_df['title'].iloc[movie_ind]
    return list_recommend
```

```python
recommend_movie('Avatar')
```

```python
recommend_movie('The Dark Knight Rises')
```
